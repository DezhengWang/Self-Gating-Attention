```python
import pandas as pd
```

```python
net_name = f'TimeXerWGA'

with open(f'20_DualAttention/result_long_term_forecast_{net_name}.txt', 'r') as f:
    res = f.readlines()
    
df = pd.DataFrame(columns=['Dataset', 'InputLen', 'PredLen', 'MSE', 'MAE', 'Repeat'])

for i in range(0, len(res), 3):
    item_title = res[i].strip().split('_sl')[-1].split('_ll')
    dataset = res[i].strip().split('long_term_forecast_')[-1].split('_')[0]
    input_len = int(item_title[0])
    pred_len = int(item_title[-1].split('_pl')[-1].split('_dm')[0])
    repeat = int(item_title[-1].lower().split('_exp_')[-1])
    item_infos = res[i+1].strip().split(',')
    mse = eval(item_infos[0].strip('mse:'))
    mae = eval(item_infos[1].strip(' mae:'))
    
    df.loc[len(df)] = [dataset, input_len, pred_len, mse, mae, repeat]

# i = 0
# df.sort_values(['Dataset', 'InputLen', 'PredLen', 'Repeat'])[i*12:(i+1) * 12]
df.groupby(['Dataset', 'InputLen', 'PredLen']).mean()#.to_csv('Ablation_TimeXerWGA.csv')
tmp = df.groupby(['Dataset', 'InputLen', 'PredLen']).mean()
tmp['MSE SD'] = df.groupby(['Dataset', 'InputLen', 'PredLen']).std()['MSE']
tmp['MAE SD'] = df.groupby(['Dataset', 'InputLen', 'PredLen']).std()['MAE']
tmp.to_csv(f'results_log/Ablation_{net_name}.csv')
```
