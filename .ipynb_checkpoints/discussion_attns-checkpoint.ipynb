```python
test_data, test_loader = exp._get_data(flag='test')

scores = []
score_res = []
score_base = []

for i in [98, 964, 966, 1200, 949]: #[98, 964, 966, 2212, 949]
    
    exp.model.eval()
    with torch.no_grad():
        batch_x, batch_y, batch_x_mark, batch_y_mark = test_data[i]

        batch_x = torch.Tensor(batch_x).unsqueeze(0).float().to(exp.device)

        batch_y = torch.Tensor(batch_y).unsqueeze(0).float().to(exp.device)
        batch_x_mark = torch.Tensor(batch_x_mark).unsqueeze(0).float().to(exp.device)
        batch_y_mark = torch.Tensor(batch_y_mark).unsqueeze(0).float().to(exp.device)

        # decoder input
        dec_inp = torch.zeros_like(batch_y[:, -exp.args.pred_len:, :]).float()
        dec_inp = torch.cat([batch_y[:, :exp.args.label_len, :], dec_inp], dim=1).float().to(exp.device)
        
        outputs = exp.model(batch_x, batch_x_mark, dec_inp, batch_y_mark)[0]
        
        score_item = exp.model.encoder.layers[0].self_attention.inner_attention.attn.cpu().detach().numpy()[0]
        scores.append(score_item)
        
        
scores = np.array(scores)

# scores = scores - alpha_w2 / (alpha_w2+alpha_w1) * scores

fig, axes = plt.subplots(scores.shape[0], 8, dpi=400)

for i in range(scores.shape[0]):
    for j in range(8):
        ax = axes[i, j]
        im = ax.imshow(scores[i, j], cmap='viridis')
        ax.axis('off')
        ax.invert_yaxis()

cbar = fig.colorbar(im, ax=axes.ravel().tolist(), orientation='vertical', 
                    fraction=0.015, pad=0.01)

plt.tight_layout(rect=[0, 0, 0.9, 0])
plt.show()

scores_res = scores

scores = [np.hstack(item) for item in scores.transpose(0,1,3,2)]
# scores[-1] = np.array(scores[:4]).mean(axis=0, keepdims=True)[0]
# alpha = scores[-1]
# pd.DataFrame(np.vstack(scores)).to_csv('TimeXer_Exchange_alpha.csv')

scores_res = scores_res[:4] - np.array(scores_res[:4]).mean(axis=0, keepdims=True)[0]
scores_res = np.abs(scores_res)
scores_res = [np.hstack(item.transpose(0,2,1)) for item in scores_res]
pd.DataFrame(np.vstack(scores_res)).to_csv('TimeXer_Exchange_res.csv')
```

# seq len analysis

```python
import numpy as np
import pandas as pd
```

```python
# baselines = ['TimeXer', 'TimeXerWAuto', 'TimeXerWGA', 'TimeXerWGeom', 'TimeXerWProb',  'TimeXerWTOST', ]
baselines = ['TimeXerWGA', ]

for baseline in baselines:
    with open(f'exp_results/seq len/logs/result_long_term_forecast_{baseline}.txt', 'r') as f:
        res = f.readlines()
        
    df = pd.DataFrame(columns=['InputLen', 'PredLen', 'MSE', 'MAE', 'train_speed', 'inference_speed', 'Repeat'])
    
    for i in range(0, len(res), 3):
        item_title = res[i].strip().split('_sl')[-1].split('_ll')
        input_len = int(item_title[0])
        pred_len = int(item_title[-1].split('_pl')[-1].split('_dm')[0])
        repeat = int(item_title[-1].lower().split('_exp_')[-1])
        item_infos = res[i+1].strip().split(',')
        mse = eval(item_infos[0].strip('mse:'))
        mae = eval(item_infos[1].strip(' mae:'))
        train_speed = eval(item_infos[2].strip(' train_speed:'))
        inference_speed = eval(item_infos[3].strip(' inference_speed:'))
        
        df.loc[len(df)] = [input_len, pred_len, mse, mae, train_speed, inference_speed, repeat]
    
    df.sort_values(['InputLen', 'PredLen', 'Repeat']) \
                    .groupby(['InputLen', 'PredLen']) \
                    .agg(['mean', 'std']).to_csv(f'exp_results/seq len/logs/results_{baseline}_seqlen.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, ifft
```

```python
# ETTh1_96_96
test_data, test_loader = exp._get_data(flag='test')

for i, (batch_x, batch_y, batch_x_mark, batch_y_mark) in enumerate(test_loader):
    if i == 200:
        break
batch_x = batch_x.float()
batch_x_mark = batch_x_mark.float()
```

```python
exp.model.load_state_dict(torch.load("checkpoints_overview/checkpoint_0.pth"))
exp.model = exp.model.to('cpu')
exp.model.eval()
outputs_0 = exp.model(batch_x, batch_x_mark, None, None)

exp.model.load_state_dict(torch.load("checkpoints_overview/checkpoint_1.pth"))
exp.model = exp.model.to('cpu')
exp.model.eval()
outputs_1 = exp.model(batch_x, batch_x_mark, None, None)

exp.model.load_state_dict(torch.load("checkpoints_overview/checkpoint_2.pth"))
exp.model = exp.model.to('cpu')
exp.model.eval()
outputs_2 = exp.model(batch_x, batch_x_mark, None, None)
```

```python
original_signal = batch_x[0][:,2].numpy()
n = len(original_signal)

signal_no_dc = original_signal - np.mean(original_signal)

yf = fft(signal_no_dc)
xf = fftfreq(n, 1)[:n//2]
amplitude = 2.0 / n * np.abs(yf[:n//2])

top_k = 2
top_indices = np.argsort(amplitude)[-top_k:][::-1]
top_freqs = xf[top_indices]
top_amps = amplitude[top_indices]

individual_periodic = []
for idx in top_indices:
    single_yf = np.zeros_like(yf, dtype=complex)
    single_yf[idx] = yf[idx] 
    single_yf[n - idx] = yf[n - idx]
    
    single_signal = np.real(ifft(single_yf)) + np.mean(original_signal)
    individual_periodic.append(single_signal)

all_periodic = np.sum(individual_periodic, axis=0)
residual = original_signal - all_periodic

pd.DataFrame({
    "Original": original_signal,
    "individual_periodic_0": individual_periodic[0],
    "individual_periodic_1": individual_periodic[1],
    # "individual_periodic_2": individual_periodic[2],
    "all_periodic": all_periodic,
    "residual": residual,
    "GrandTrue": batch_y[0, -96:, 2].detach().numpy(),
    "PredictY_1": outputs_0[0, :, 2].detach().numpy(),
    "PredictY_2": outputs_1[0, :, 2].detach().numpy(),
    "PredictY_3": outputs_2[0, :, 2].detach().numpy(),
}).to_csv("decomposition_etth1_ot_200.csv")
```
